## Manipulating TaiChung DE Power Data 

This data from TaiChung DE PV Plant

2016/01/01-2017/12/31 

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# quick look data 
power = pd.read_csv('data\\power.csv', parse_dates=['Date'], index_col='Date')
power.head(12)
```

```python
# use 0 fill nan
power_fill_zero = power.fillna(0, inplace=False)
power_fill_zero.head()
```

```python
# Visualize the first fifteen days
plt.style.use('fivethirtyeight')
ax1 = power_fill_zero.loc['2016-01-01':'2016-01-15','power(KWH)'].plot(figsize=(16,9), fontsize=12, linewidth=3, linestyle='-')
ax1.set_xlabel('Date', fontsize=12)
ax1.set_ylabel('Power(KWH)', fontsize=12)
ax1.set_title('TaiChung DE 2016/01/01 - 2016/01/11',fontsize=16)
plt.show()
```

```python
# aggregrate every hour mean value
power_fill_zero['hour'] = power_fill_zero.index.hour
power_grouped = power_fill_zero.groupby('hour')
power_mean_by_hour = power_grouped.mean()
ax2 = power_mean_by_hour.plot(figsize=(12,8), fontsize=12, linewidth=3, linestyle='-',xticks=range(0,24,2))
ax2.set_xlabel('Hour', fontsize=12)
ax2.set_ylabel('Power(KWH)', fontsize=12)
ax2.set_title('Power Mean by Hour',fontsize=16)
plt.show()
```

```python
# box plot every hour
ax3 = power_fill_zero.boxplot(column='power(KWH)',by='hour', figsize=(12,8), fontsize=12)
ax3.set_xlabel('Hour', fontsize=12)
ax3.set_ylabel('Power(KWH)', fontsize=12)
plt.show()
```

```python
# replace outlier
def replace_outlier(series):
    Q1 = series.quantile(.25)
    Q3 = series.quantile(.75)
    IQR = (Q3-Q1)
    upper = Q3 + 1.5 * IQR
    lower = Q1 - 1.5 * IQR
    series = np.where(series > upper, upper, np.where(series < lower, 
                      lower, series))
    return series

power_fill_zero['power(KWH)_replaced'] = power_fill_zero['power(KWH)']

for i in range(6,20,1):
    hour_idx = (power_fill_zero.hour == i)
    power_hour = power_fill_zero.loc[hour_idx, ['power(KWH)']]
    power_fill_zero.loc[hour_idx,['power(KWH)_replaced']] = replace_outlier(power_hour)

ax4 = power_fill_zero.boxplot(by='hour',column='power(KWH)_replaced', figsize=(12,8), fontsize=12)
ax4.set_xlabel('Hour', fontsize=12)
ax4.set_ylabel('Power(KWH)', fontsize=12)
plt.show()
```

```python
# save as csv
power_fill_zero.to_csv('data\\power_no_outlier.csv', encoding='utf-8-sig')
```
