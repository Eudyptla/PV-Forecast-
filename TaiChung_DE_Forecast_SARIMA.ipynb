# Test for stationarity and seasonality

4 questions:
    
    1.Is the time series stationary?
    
    2.What differecing make it stationary?
    
    3.What transformers will make it stationary?
    
    4.What values of p and q are most promising?

Try Dicky-Fuller test.

Null hypothesis is non-stationary.

```python
# Dicky-Fuller test
results = adfuller(power)
#Perform Dickey-Fuller test result:
print('Results of Dickey-Fuller Test:')
results_output = pd.Series(results[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])

for key,value in results[4].items():
    results_output['Critical Value (%s)'%key] = value

print (results_output) 
```

Test statistic : -12.03644629189143 → more negative means more likely to be stationary.

P-value: 2.796121583625262e-22
    
If p-value is small → reject null hypothesis. Reject non-stationary.

Try take diff(1).


```python
# take diff(1) and dropna
power_diff = power.diff(1).dropna()
# Dicky-Fuller test
results = adfuller(power_diff)
#Perform Dickey-Fuller test result:
print('Results of Dickey-Fuller Test:')
results_output = pd.Series(results[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])

for key,value in results[4].items():
    results_output['Critical Value (%s)'%key] = value

print (results_output) 
```

P-value = 0 . **The time series is stationary.**

time series = trend + seasonal + residual

Seasonal data has predictable and repeated patterns.

This is hourly data. period =24

Look head ten days.

```python
decomp_results = seasonal_decompose(power_diff,period=24)
fig,(ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize = (12, 16))
decomp_results.observed[:240].plot(ax=ax1, legend=False)
ax1.set_ylabel('Observed')
decomp_results.trend[:240].plot(ax=ax2, legend=False)
ax2.set_ylabel('Trend')
decomp_results.seasonal[:240].plot(ax=ax3, legend=False)
ax3.set_ylabel('Seasonal')
decomp_results.resid[:240].plot(ax=ax4, legend=False)
ax4.set_ylabel('Residual')

plt.show()
```

# Find promising model orders

Seasonal ARIMA = SARIMA

SARIMA(p,d,q)(P,D,Q,S)

* Non-seasonal orders
    
        p:autoregrassive order
    
        d:differencing order
    
        q:moving average order

* Seasonal orders
    
        P:Seasonal autoregrassive order
    
        D:Seasonal differencing order
    
        Q:Seasonal moving average order
    
        S:number of time steps per cycle
    

Plot the ACF and PACF charts and find the optimal parameters

ACF - Autocorrelation function.

PACF - Partial autocorrelation function.


Name|AR(P)              |MA(q)              |ARMA(p,q)
----|:-----------------:|:-----------------:|:----:
ACF |Tails off          |cut off after lag q|Tails off  
PACF|cut off after lag p|Tails off          |Tails off  



```python
# Take the seasonal differences and drop NaNs
power_seasonal = power_diff.diff(24).dropna()
# finding p,q
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(12,8))
# Make ACF plot up to 23
plot_acf(power_seasonal, lags=23, zero=False, ax=ax1)
# Make PACF plot up to 23
plot_pacf(power_seasonal, lags=23, zero=False, ax=ax2)
plt.show()
```

ACF :not tails off and cut off
    
PACF:not tails off and cut off
    
$\therefore p = 0, q = 0$


```python
# Make a list of the first 10 seasonal lags
lags = [24, 48, 72, 96, 120, 144, 168, 192, 216, 240]
# finding P, Q
# Create figure
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(12,8))
# Make ACF plot 
plot_acf(power_seasonal, lags=lags, zero=False, ax=ax1)
# Make PACF plot
plot_pacf(power_seasonal, lags=lags, zero=False, ax=ax2)
plt.show()
```

ACF: cut off after lag 1.

PACF: tails off.

$\therefore P=0, Q=1$
  
Based on the ACF and PACF plots

We get SARIMA(0,1,0)(0,1,1,24).

$$y^{\prime}_{t} = m_{24}\epsilon_{t-24} + \epsilon_{t} $$

# Fit models and narrow selection with AIC/BIC

```python
# Instantiate model
model = SARIMAX(power, order=(0, 1, 0), seasonal_order=(0, 1, 1, 24))
# Fit model
results = model.fit()
# Print fit summary
print(results.summary())
```

AIC and BIC 

AIC - Akaike information criterion. Lower AIC indicates a better model.

BIC - Bayesian information criterion. Lower BIC indicates a better model.

# Perform model diagnostics tests

Plot diagnostics.

```python
results.plot_diagnostics(figsize=(16,16))
plt.show()
```

# Visualizing Results

```python
# predict
predict = results.get_prediction()
predict_ci = predict.conf_int()
# Graph
fig, ax = plt.subplots(figsize=(16,6))
npre = 4
ax.set(title='SARIMA Predict', xlabel='Date', ylabel='power(KWH)')

# Plot data points
power['2017-09-01':].plot(ax=ax, style='o', label='Actual')

# Plot predictions
predict.predicted_mean.loc['2017-09-01':].plot(ax=ax, style='r--', label='Predict')
ci = predict_ci.loc['2017-09-01':]
ax.fill_between(ci.index, ci.iloc[:,0], ci.iloc[:,1], color='r', alpha=0.1)

ax.set_ylim(-200, 1500)
legend = ax.legend(loc='lower right')
```

```python
power_test = data.loc['2017-09-01':,['power(KWH)']]
power_test['Predicted'] = predict.predicted_mean['2017-09-01':]
daytime_idx = (power_test.index.hour) >= 6 & (power_test.index.hour <= 18)
predict_test = power_test[daytime_idx]
# Predict error 
fig, ax = plt.subplots(figsize=(16,6))
npre = 4
ax.set(title='Predict error', xlabel='Date', ylabel='Predict - Actual')
# In-sample predict and 95% confidence intervals
predict_test['error'] = predict_test['Predicted'] - predict_test['power(KWH)']
predict_test['error'].plot(ax=ax, label='Prediction')
ci = predict_ci.loc['2017-09-01':].copy()
ci_day = ci[daytime_idx]
ci.iloc[:,0] -= predict_test['power(KWH)']
ci.iloc[:,1] -= predict_test['power(KWH)']
ax.fill_between(ci.index, ci.iloc[:,0], ci.iloc[:,1], alpha=0.1)

legend = ax.legend(loc='lower left');
legend.get_frame().set_facecolor('w')
ax.set_ylim(-1000, 1000)
```

# Model Score 

```python
# Compute and print the reslut
def MRE(x,y):
    return np.nanmean(np.abs((x-y)/np.max(x)))

def print_results(y_test, y_predict):
    print('Root mean squared error:{:.4f}'.format(np.sqrt(mean_squared_error(y_test,y_predict))))
    print('Mean absolute error:{:.4f}'.format(mean_absolute_error(y_test,y_predict)))
    print('Mean relative error:{:.2%}'.format(MRE(y_test,y_predict)))
    print('Coefficient of determination:{:.4f}'.format(r2_score(y_test,y_predict)))

print_results(predict_test['power(KWH)'], predict_test['Predicted'])
```

# Look at Worst and Best Predicted Days

```python
predict_test['abs_error'] = predict_test['error'].apply(np.abs)
predict_test['year']=predict_test.index.year
predict_test['month']=predict_test.index.month
predict_test['day']=predict_test.index.day

error_by_day = predict_test[:'2017-12-31'].groupby(['year','month','day']).mean()[['power(KWH)','Predicted','error','abs_error']]

def look_ten(data, Best=True):
    top_ten=data.sort_values('abs_error', ascending=Best).head(10)
    top_date = top_ten.index[0]
    top_date = str(top_date[0])+'-'+str(top_date[1])+'-'+str(top_date[2])
    print(top_ten)
    return top_date
```

Looking worst predicted ten days.

```python
# Worst absolute predicted days
Worst_date = look_ten(error_by_day, Best=False)
```

Looking best predicted ten days.

```python
# Best predicted days
Best_date = look_ten(error_by_day, Best=True)
```
