## Taichung DE forecast by LSTM

Data from TaiChung DE PV plant, CWB WUQI weather.

Collect 2016/01/01-2017/12/31.

Time frequcy: 1 hour.

Try LSTM to forecast.

```python
# import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import TimeSeriesSplit, train_test_split, cross_val_score
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, LSTM, TimeDistributed, RepeatVector
from keras.callbacks import EarlyStopping
```

```python
# quick look data
data = pd.read_csv('data\\Taichung_Power_Weather.csv', parse_dates=['Date'], index_col='Date')
# visualizing data
data.plot(subplots=True, linewidth=0.5, layout=(8, 2), figsize=(12, 16),
          sharex=False, sharey=False)
plt.show()
```

```python
# replace outlier
def replace_outlier(series):
    series_mean = series.mean()
    series_std = series.std()
    upper = series_mean + 3 * series_std 
    lower = series_mean - 3 * series_std
    series_new = np.where(series > upper, upper, np.where(series < lower, lower, series))
    return series_new
# choose day time
daytime_idx = ((6<= data.index.hour) & (data.index.hour<= 18))

for col in data.columns:
    data.loc[daytime_idx,col] = replace_outlier(data.loc[daytime_idx,col])

print(data[daytime_idx].describe().transpose())
```

Drop WS, WD, WDGust and WDGust because this is PV forecast.

Drop Precp and PrecpHour ,because 75% data are zero.

```python
data_drop = data.drop(columns=['WS', 'WD', 'WSGust', 'WDGust', 'Precp', 'PrecpHour'])
```

Norm min =-1, max =1.

```python
X_scale = MinMaxScaler(feature_range=(-1, 1))
y_scale = MinMaxScaler(feature_range=(-1, 1))
X_norm = X_scale.fit_transform(data_drop.iloc[:,1:])
y_norm = y_scale.fit_transform(data_drop.iloc[:,[0]])
```

```python
def build_data(X_norm, y_norm, pasttime, futuretime):
    X, y = [], []
    for i in range(data.shape[0]-futuretime-pasttime):
        X.append(np.array(X_norm[i:i+pasttime]))
        y.append(np.array(y_norm[i+pasttime:i+pasttime+futuretime,0]))
    
    return np.array(X), np.array(y)
X, y = build_data(X_norm, y_norm, 24, 1)
```

```python
# split train test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.166,shuffle=False)
```

```python
def buildManyToOneModel(shape):

    model = Sequential()
    model.add(LSTM(64, input_length=shape[1], input_dim=shape[2], activation='relu', return_sequences=True))
    model.add(LSTM(32, activation='relu'))
    model.add(Dense(1))
    model.compile(loss="mean_squared_error", optimizer="adam"
                ,metrics=['mean_squared_error'])
    return model

model = buildManyToOneModel(X_train.shape)
print(model.summary())
```

```python
callback = EarlyStopping(monitor="val_loss", patience=10, verbose=1, mode="auto")
history = model.fit(X_train, y_train, epochs=1000, batch_size=128, validation_data=(X_test, y_test), callbacks=[callback])
```

```python
def plot_history(history):
    hist = pd.DataFrame(history.history)
    hist['epoch'] = history.epoch
    plt.figure()
    plt.xlabel('Epoch')
    plt.ylabel('Mean Square Error')
    plt.plot(hist['epoch'], hist['mean_squared_error'],
             label='Train Error')
    plt.plot(hist['epoch'], hist['val_mean_squared_error'],
             label='Val Error')
    plt.legend()

# The patience parameter is the amount of epochs to check for improvement

plot_history(history)
```

```python
y_test_value = y_scale.inverse_transform(y_test)
y_predict_norm = model.predict(X_test)
y_predict = y_scale.inverse_transform(y_predict_norm)
```

```python
# visualizing predict
fig, ax = plt.subplots(figsize=(18,12))
ax.plot(y_test_value, color='k', alpha=.3, lw=3)
ax.plot(y_predict)
ax.set(xlabel="Time")
plt.show()
```

```python
def MRE(x,y):
    return np.nanmean(np.abs((x-y)/np.max(x)))

print('Mean squared error:{:.4f}'.format(mean_squared_error(y_test_value,y_predict)))
print('Mean absolute error:{:.4f}'.format(mean_absolute_error(y_test_value,y_predict)))
print('Mean relative error:{:.2%}'.format(MRE(y_test_value,y_predict)))
print('Coefficient of determination:{:.4f}'.format(r2_score(y_test_value,y_predict)))
```
